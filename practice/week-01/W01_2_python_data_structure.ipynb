## 1. List

```python
# List 선언 및 초기화

x = [] # 또는 list()로도 생성할 수 있습니다.
x = [1,'a',True] 
print(x)
print(type(x))
```

```python
# List 원소 접근 (READ)
x = [1,2,3] 
print(x[0])
print(type(x[0]))
```

```python
# List 원소 추가 (WRITE)
x = [1,2,3] 
print(x)
x.append(4)
print(x)
```

```python
# List 원소 변경 (UPDATE)
x = [1,2,3,4] 
print(x)
x[2] = 100
print(x)
```

```python
# List 원소 삭제 (DELETE)
x = [1,2,3,4] 
print(x)
del x[2]
print(x)
```

```python
# 원소 존재여부 확인
1 in x
```

## 2. Dictionary

```python
x = {}  # x = dict()
print(x)
print(type(x))
```

```python
x = {'a':1, 'b':2, 'c':3}  # x = dict()
print(x)
print(x['a'])
```

```python
x['d']=4
print(x)
```

```python
x['d']=10
print(x)
```

```python
del x['d']
print(x)
```

```python
'd' in x
```

```python
x = {'a':3, 'b':2, 'c':1}
print(x.keys())
print(x.values())
print(x.items())
```

```python
sorted_x = sorted(x.items(), key=lambda kv: kv[0], reverse=True)
print(sorted_x)
```

```python
sorted_x = sorted(x.items(), key=lambda kv: kv[1])
print(sorted_x)
```

```python
# Q. Value가 오름차순으로 정렬하되, Value가 같을때는 Key를 기준으로 정렬 하려면?
x = {'a':3, 'b':2, 'c':1, 'd':3}
sorted_x = sorted(x.items(), key=lambda kv: (kv[1], kv[0]))
print(sorted_x)
```

## 3. Set

```python
x = {1,2,3} # set()
type(x)
```

```python
# READ
for a in x:
    print(a)
```

```python
# INSERT
x.add(1)
x.add(4)
x.add(5)
x.add(6)
print(x)
```

```python
# UPDATE
x.update([1,2,7])
print(x)
```

```python
# DELETE
x.remove(7)
print(x)
```

```python
a = {1,2,3,4,5}
b = {4,5,6,7,8}
#교집합
print(b.intersection(a))
print(a.intersection(b))
#합집합
print(a.union(b))
print(b.union(a))
#차집합
print(a.difference(b))
print(b.difference(a))

```

```python
x = ['Korea', 'Canada', 'Japan', 'USA', 'Korea', 'Italy']
print(x, type(x))
x = set(x)
print(x, type(x))
```

## 5. 자료구조간의 타입 캐스팅(casting)

```python
a = [1,2,3,4]
print(a, type(a))
set_a = set(a)
print(set_a, type(set_a))
tuple_a = tuple(a)
print(tuple_a, type(tuple_a))

dict_a = dict(a)
print(dict_a, type(dict_a))
```

```python
a = {'a':1, 'b':2, 'c':3}
print(a, type(a))
set_a = set(a)
print(set_a, type(set_a))
list_a = list(a)
print(list_a, type(list_a))
tuple_a = tuple(a)
print(tuple_a, type(tuple_a))

list_a = list(a.items())
set_a = set(a.items())
tuple_a = tuple(a.items())
print(list_a)
print(set_a)
print(tuple_a)
```
